```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
dir_non_agri = './images_dataSAT/class_0_non_agri/'
dir_agri = './images_dataSAT/class_1_agri/'
```

```python
non_agri_images_paths = sorted(glob.glob(os.path.join(dir_non_agri, '*')))
image_data = np.array(Image.open(non_agri_images_paths[0]))
```

## Task 1: Determine the shape (dimensions) of a single image stored in the image_data variable.

```python
image_data.shape
```

## Task 2: Display the first four images in the './images_dataSAT/class_0_non_agri/' directory.

```python
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for i in range(4):
    img = np.array(Image.open(non_agri_images_paths[i]))
    axes[i].imshow(img)
    axes[i].set_title(os.path.basename(non_agri_images_paths[i]))
    axes[i].axis('off')
plt.tight_layout()
plt.show()
```

## Task 3: Create a list named agri_images_paths that contains the full file paths of all images located in the dir_agri directory. Sort the list before saving it.

```python
agri_images_paths = sorted(glob.glob(os.path.join(dir_agri, '*')))
agri_images_paths[:5]
```

## Task 4: Determine the number of images of agricultural land that exist in the './images_dataSAT/class_1_agri/' directory.

```python
len(agri_images_paths)
```

## Task 5: Display the first four images of the agricultural land.

```python
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for i in range(4):
    img = np.array(Image.open(agri_images_paths[i]))
    axes[i].imshow(img)
    axes[i].set_title(os.path.basename(agri_images_paths[i]))
    axes[i].axis('off')
plt.tight_layout()
plt.show()
```
