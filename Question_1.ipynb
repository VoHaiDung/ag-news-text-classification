```python
# Import necessary libraries
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Define directories
dir_non_agri = './images_dataSAT/class_0_non_agri/'
dir_agri = './images_dataSAT/class_1_agri/'
```

```python
# Get sorted list of non-agricultural image paths
non_agri_images_paths = sorted(glob.glob(os.path.join(dir_non_agri, '*')))

# Load the first image into image_data
image_data = np.array(Image.open(non_agri_images_paths[0]))
print("Image loaded successfully.")
```

## Task 1: Determine the shape (dimensions) of a single image stored in the image_data variable.

```python
# Task 1: Determine the shape of the image
print("Shape of image_data:", image_data.shape)
```

## Task 2: Display the first four images in the './images_dataSAT/class_0_non_agri/' directory.

```python
# Task 2: Display first 4 non-agricultural images
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for i in range(4):
    img = Image.open(non_agri_images_paths[i])
    axes[i].imshow(img)
    axes[i].set_title(f"Non-Agri Image {i+1}")
    axes[i].axis('off')
plt.suptitle("First 4 Non-Agricultural Images", fontsize=14)
plt.tight_layout()
plt.show()
```

## Task 3: Create a list named agri_images_paths that contains the full file paths of all images located in the dir_agri directory. Sort the list before saving it.

```python
# Task 3: Create sorted list of agricultural image paths
agri_images_paths = sorted(glob.glob(os.path.join(dir_agri, '*')))

# Verify by printing the first 5 paths
print("First 5 agricultural image paths:")
for p in agri_images_paths[:5]:
    print(p)
print(f"\nTotal paths stored: {len(agri_images_paths)}")
```

## Task 4: Determine the number of images of agricultural land that exist in the './images_dataSAT/class_1_agri/' directory.

```python
# Task 4: Count agricultural images
num_agri_images = len(agri_images_paths)
print(f"Number of agricultural land images: {num_agri_images}")
```

## Task 5: Display the first four images of the agricultural land.

```python
# Task 5: Display first 4 agricultural images
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for i in range(4):
    img = Image.open(agri_images_paths[i])
    axes[i].imshow(img)
    axes[i].set_title(f"Agri Image {i+1}")
    axes[i].axis('off')
plt.suptitle("First 4 Agricultural Land Images", fontsize=14)
plt.tight_layout()
plt.show()
```
